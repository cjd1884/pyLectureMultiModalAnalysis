```python
import pandas as pd
```

```python
df = pd.read_csv('index.csv', sep=';')
df
```

```python
classes = df['CLASS_1'].unique()
classes
```

```python
import os
import shutil
for cl in classes:
    if not os.path.exists(cl):
        os.makedirs(cl)
```

```python
cnt_b=0
cnt_i=0
cnt_n=0
for f,s,c in df[['FILE','SEG','CLASS_1']].values:
      
        pathOut='video/'+f+'/'+s+'.mp4'
        print(pathOut,c)

        if c == 'boring':
            cnt_b=cnt_b+1
            new=c+'/'+f+'_'+s+'.mp4'
            print(new)
            shutil.move(pathOut,new )
        if c == 'interesting':
            cnt_i=cnt_i+1
            new=c+'/'+f+'_'+s+'.mp4'
            print(new)
            shutil.move(pathOut,new )
        if c == 'neutral':
            cnt_n=cnt_n+1
            new=c+'/'+f+'_'+s+'.mp4'
            print(new)
            shutil.move(pathOut,new )
```
